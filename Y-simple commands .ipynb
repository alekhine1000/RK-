```python
# Rename columns
df_new = df.rename(columns={'Product':'PRODUCT','Price':'PRICE'}) 
```

```python
df_new.head()
```

```python
#drop columns
df_new_1=df_new.drop(columns=['PRODUCT'])
```

```python
df_new_1.head()
```

```python
df_new_2 = df_new_1.copy()  # âœ… Correctly create a copy

# Convert 'PRICE' column by removing '$' and ',' and converting to float
df_new_2['PRICE'] = df_new_2['PRICE'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Convert 'Discount' column by removing '%' and converting to int
df_new_2['Discount'] = df_new_2['Discount'].replace({'%': ''}, regex=True).astype(int)
```
